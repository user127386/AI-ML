```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/students.csv") #reading the csv files to operate
```

```python
df.info()#shows csv files information on how many rows coloumn and values types
```

```python
print(df.isnull().sum())#if threre are null then sum up them and how many
```

```python
df = df.dropna() #deletes all empty row
```

```python
df[df["grade"]==11] #if student is at 11th grade
```

```python
#if there fav sub is math
df_fav = [df["favorite_subject"] == "Math"]
print(df_fav)
```

```python
#to remove the commas which might have been mistyped in the name coloumn
df['name'] = df['name'].astype(str).str.replace(',','', regex = False)
```

```python
print(df['name'])
```

```python
#to fill the age coloumn to let no null value exist
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
print(df.round(2))
```

```python
#changing the score of student which was typed in words
#and finding median to fill the score coloumn using median
df['score'] = df['score'].replace("Eighty Two",82)
df['score'] = df['score'].fillna(df['score'].median())
print(df.round(2))
```

```python
#check whether a row has been repeated
dupe_count = df.duplicated().sum()
print(dupe_count)
```

```python
#deletes the repeated row
df_clean = df.drop_duplicates()
print(df_clean)
```

```python
#to safely parse the strings to numerical value
df_clean['score'] = pd.to_numeric(df_clean['score'], errors='coerce')
print(df_clean)
```
