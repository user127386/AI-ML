```python
#senors readings
sensors = np.array([
    [12.1, 14.3, 15.2], #sensors 1
    [10.2, 11.4, 14.6], #sensors 2
    [19.5, 20.3, 11.4], #sensors 3
    [22.3, 21.3, 14.4]  #sensors 4
])
```

```python
specific_reading = sensors[2,1]
print(specific_reading)
```

```python
top_readings = sensors[0:2,:]
print(top_readings)
```

```python
final_readings = sensors[:, 2]
print(final_readings)
```

```python
flat_temps = np.array((100, 90, 80, 98, 110, 100, 101, 110, 105, 104, 103, 102))
structured_temps = flat_temps.reshape(4, 3)
print("Structured Temperatures: \n",structured_temps)
```

```python
celcius_temperature = ((structured_temps*32 - 32) * 5/9)
print("In celsius: \n", celcius_temperature.round(2))
```

```python
score = np.array([
    [85, 95, 99],
    [13, 45, 50],
    [88, 91, 92]
])
print(score.min())
print(score.max())
print(score.mean().round(2))
```

```python
import time
size = 1_000_000
python_list = list(range(size))
numpy_array = np.arange(size)
```

```python
start_time = time.time()
python_list_result = [x*2 for x in python_list]
python_duration = time.time() - start_time
```

```python
start_time= time.time()
numpy_result = numpy_array*2
numpy_duration = time.time() - start_time
```

```python
print(f"time taken by standard python list:{python_duration} seconds")
print(f"time taken by nummpy list: {numpy_duration} seconds")
print(f" numpy is roughly {round(python_duration/numpy_duration)} seconds faster")
```
