```python
def calculate_tax (amount, tax_rate = 0.08):
  sales_tax = amount * tax_rate
  return round(sales_tax, 2)
```

```python
item_price = 125.0
tax_due = calculate_tax(item_price)
print(f"Tax on $(item_price) $(tax_due)")
```
