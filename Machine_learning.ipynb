```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv("/content/regression_training_dataset1_5537be90-411f-4a6b-8ea4-7bb197181da5_268917_.csv")
```

```python
df.info()
```

```python
x = df.drop(columns = "PerformanceScore")
y = df["PerformanceScore"]
```

```python
print(y)
```

```python
y_stratified_bins = pd.qcut(y, q=5, labels= False)
x_train_val,x_test,y_train_val, y_test = train_test_split(
    x,y,
    test_size=0.2,
    random_state = 40,
    stratify = y_stratified_bins
)
```

```python
y_train_val_bins = pd.qcut(y_train_val, q=5, labels=False)
x_train, x_val, y_train, y_val = train_test_split(
    x_train_val, y_train_val,
    test_size=0.25,
    random_state=40,
    stratify = y_train_val_bins
)
```

```python
print(x.shape)
print(x_val.shape)
print(y_val.shape)
print(x.head(2))
print(x_val.head(2))
```

```python
standard_scaler = StandardScaler()
standard_scaler.fit(x_train)
```

```python
x_train_standardization = standard_scaler.transform(x_train)
x_val_scaled_std = standard_scaler.transform(x_val)
x_test_scaled_std = standard_scaler.transform(x_test)
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(x_train)
x_train_scaled_norm = min_max_scaler.transform(x_train)
x_val_scaled_norm = min_max_scaler.transform(x_val)
x_test_scaled_norm = min_max_scaler.transform(x_test)
```
