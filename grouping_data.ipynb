```python
import numpy as np
import pandas as pd
```

```python
df_sales = pd.read_csv("/content/advanced_pandas_sales_dataset (1).csv")
```

```python
df_sales.info()
```

```python
category_revenue = df_sales.groupby('Category')['Revenue'].sum().reset_index
```

```python
print(category_revenue)
```

```python
region_metrics = df_sales.groupby('Region').agg(
    Total_Units_sold=('Quantity', "sum"),
    Average_Items=('UnitPrice', "mean"),
).reset_index
```

```python
print(region_metrics)
```

```python
managers_data = {
    "Region": ["North", "South", "East", "West"],
    "Manager": ["Alice", "Bob", "Charlie", "David"]
}
df_managers = pd.DataFrame(managers_data)
```

```python
df_enriched = pd.merge(df_sales, df_managers, on='Region', how = 'left')
```

```python
print(df_enriched.head(3))
```

```python
revenue_pivot = pd.pivot_table(
    df_sales,
    values= 'Revenue',
    index= 'Region',
    columns= 'Category',
    aggfunc= 'sum',
)
print(revenue_pivot)
```

```python
order_count = pd.crosstab(
    index = df_sales['Region'],
    columns = df_sales['Category'],
)
```

```python
print(order_count)
```

```python
print(df_sales['OrderDate'].dtype)
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'])
print(df_sales['OrderDate'].dtype)
```

```python
df_sales['Month_Name'] = df_sales['OrderDate'].dt.strftime("%B")
df_sales['Day_of_the_week'] = df_sales['OrderDate'].dt.strftime("%A")
monthly_trends = df_sales.groupby(df_sales['OrderDate'].dt.to_period('M'))['Revenue'].sum()
```

```python
print(monthly_trends)
```

```python
df_election = pd.read_csv("/content/election_data.csv")
df_province = pd.read_csv("/content/province_data.csv")
df_district = pd.read_csv("/content/district_data (1).csv")
```

```python
print(df_election['candidate_id'].head(3))
```

```python
print(df_province.head(3))
```

```python
print(df_district.head(3))
```

```python
df_candidate_districts = pd.merge(
    df_election,
    df_district,
    on='district_id',
    how='inner'
)
```

```python
print(df_candidate_districts.shape)
```

```python
print(df_candidate_districts[['candidate_id','candidate_name', 'party']].head(3))
```

```python
df_district_left = pd.merge(
    df_district,
    df_province,
    on='province_id',
    how='left'
)
```

```python
print(df_district_left[['district_name','province_name']].head(3))
```

```python
df_master_election = pd.merge(
    df_election,
    df_district,
    on = 'district_id',
    how= 'inner'
).merge(
    df_province,
    on='province_id',
    how='inner'
)
```

```python
print(df_master_election[['candidate_id','candidate_name', 'age', 'gender']].head(3))
```
