```python
import numpy as nm
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("/content/regression_training_dataset1_5537be90-411f-4a6b-8ea4-7bb197181da5_268917_.csv")
df.info()
```

```python
x_simple = df[['StudyHoursPerWeek']]
y_target = df[['PerformanceScore']]
```

```python
model_simple = LinearRegression()
```

```python
model_simple.fit(x_simple, y_target)
```

```python
print(model_simple.intercept_)
print(model_simple.coef_[0])
print(model_simple.score(x_simple, y_target))
```

```python
#Multiple feature
x_multi = df[['StudyHoursPerWeek', 'AttendancePercent', 'AssignmentsCompleted','ExperienceYears']]
model_multi = LinearRegression()
model_multi.fit(x_multi, y_target)
```

```python
print(x_multi.head(2))
```

```python
y_prediction= model_multi.predict(x_multi)
computed_mse = mean_squared_error(y_target,y_prediction)
computed_r2 = r2_score(y_target,y_prediction)
print(computed_mse)
print(computed_r2)
```
