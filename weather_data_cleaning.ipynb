```python
import pandas as pd
import numpy as np
```

```python
df_clean = pd.read_csv('/content/weather.csv')
```

```python
df_clean['humidity_percent'] = df_clean['humidity_percent'].fillna(df_clean['humidity_percent'].median())
print(df_clean)
```

```python
df_clean["city"] = df_clean["city"].fillna("Unknown")
df_clean['temperature_c'] = df_clean['temperature_c'].fillna(df_clean['temperature_c'].median())
```

```python
print(df_clean)
```

```python
df_clean = df_clean.dropna()
```

```python
df_clean['date'] = pd.to_datetime(df_clean['date'])
```

```python
print(df_clean)
```

```python
print(df_clean[['date']].dtypes)
```

```python
df_clean['city'] = df_clean['city'].str.strip()
```

```python
print(df_clean)
```

```python
df_clean['condition'] = df_clean['condition'].str.upper()
```

```python
print(df_clean)
```
