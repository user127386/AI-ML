```python
df_election = pd.read_csv("/content/election_data.csv")
df_province = pd.read_csv("/content/province_data.csv")
df_district = pd.read_csv("/content/district_data (1).csv")
```

```python
df_master = pd.merge(df_election, df_district, on='district_id', how='inner').merge(df_province, on='province_id', how = 'left')
```

```python
plt.scatter(df_master['age'], df_master['votes'], color = 'blue', alpha = 0.5)
#sns scatterplot
plt.title("Candidate Age Vs Votes", fontweight = 'bold')
plt.xlabel("Age")
plt.ylabel("Votes")
plt.grid(True, linestyle = ':', alpha = 0.6)
plt.savefig("Candidate Age VS Votes.png")
plt.close()
```

```python
party_count = df_master['party'].value_counts().head(5).sort_values(ascending=True)
plt.barh(party_count.index, party_count.values, color ='grey')
plt.title("Party Count Vs Parties")
plt.xlabel("Party Count")
plt.ylabel("Parties")
```

```python
party_count = df_master['party'].value_counts().head(5).sort_values(ascending=True)
plt.bar(party_count.index, party_count.values, color ='grey')
plt.title("Party Count Vs Parties")
plt.xticks(rotation = "vertical")
plt.ylabel("Party Count")
plt.xlabel("Parties")
```

```python
gender_distribution = df_master['gender'].value_counts()
gender_distribution
```

```python
plt.pie(
    gender_distribution.values,
    labels = gender_distribution.index,
    autopct = '%1.1f%%',
    colors = ['indigo', 'blue', 'black'],
    startangle = 140,
    pctdistance= 0.9,
    explode = [0,0,0.1]
)
```

```python
plt.hist(df_master['age'], bins = 15, color = 'indigo', edgecolor = 'black')
plt.title("Age distribution of candidates age and no of candidates")
plt.xlabel("Age")
plt.ylabel("No. Of Candidates")
plt.grid(True, alpha = 0.15)
```

```python
province_order = df_master.groupby("province_name")["age"].median().sort_values().index
```

```python
print(province_order)
```

```python
sns.set_theme(style='whitegrid')
sns.boxplot(
    data= df_master,
    x='age',
    y='province_name',
    hue = 'province_name',
    order = province_order,
    palette = "Set3"
)
plt.title("Province Age Distribution")
plt.xlabel("Age")
plt.ylabel("Province Name")
plt.savefig("provinceagedistrubution.png")
plt.close()
```
